```python
def minimax(node, depth, is_maximizing, tree):
 # If it's a leaf node (no children)
    if node not in tree or depth == 0:
        return node # node is assumed to be a number (value) at leaves
    if is_maximizing:
        best = float('-inf')
        for child in tree[node]:
            value = minimax(child, depth- 1, False, tree)
            best = max(best, value)
        return best
    else:
        best = float('inf')
        for child in tree[node]:
            value = minimax(child, depth- 1, True, tree)
            best = min(best, value)
        return best
```

```python
 game_tree = {
 'A': ['B', 'C'],
 'B': ['D', 'E'],
 'C': ['F', 'G'],
 'D': [3, 5],
 'E': [6, 9],
 'F': [1, 2],
 'G': [0,-1]
 }
```

```python
best_value = minimax('A', depth=5, is_maximizing=True, tree=game_tree)
```

```python
print("Best value for MAX:", best_value)
```

```python
# to understand how the recursion works.
def multi(A):
    if A == 1:
        return 1
    return A * multi(A - 1)
```
